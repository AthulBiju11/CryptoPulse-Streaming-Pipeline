```sql
-- price_stats
SELECT 
    window_end, 
    symbol, 
    avg_price
FROM cryptopulse_catalog.gold.dm_crypto_price_stats
WHERE
    window_end >= now() - INTERVAL '24' HOUR
ORDER BY window_end ASC;
```

```sql
SELECT
    side,
    count(*) AS trade_count,
    sum(trade_size) AS total_volume
FROM
    cryptopulse_catalog.silver.fact_crypto_trades
WHERE
    event_time >= now() - INTERVAL '1' HOUR
GROUP BY
    side;
```

```sql
SELECT
    symbol,
    price
FROM
    cryptopulse_catalog.silver.fact_crypto_trades
QUALIFY
    ROW_NUMBER() OVER (PARTITION BY symbol ORDER BY event_time DESC) = 1
```

```sql
SELECT 
    date_trunc('minute', event_time) AS trade_minute,
    symbol,
    count(*) AS num_trades,
    sum(trade_size) AS total_volume
FROM 
    cryptopulse_catalog.silver.fact_crypto_trades
WHERE 
    event_time >= now() - INTERVAL '1' HOUR
GROUP BY 
    1, 2
ORDER BY 
    1 ASC;
```

```sql
WITH latest_100_records AS (
    SELECT
        ingestion_time,
        load_timestamp
    FROM
        cryptopulse_catalog.silver.fact_crypto_trades
    ORDER BY
        event_time DESC
    LIMIT 100
)
SELECT
    avg(timestampdiff(SECOND, ingestion_time, load_timestamp)) as avg_latency_of_last_100_trades
FROM
    latest_100_records;
```
